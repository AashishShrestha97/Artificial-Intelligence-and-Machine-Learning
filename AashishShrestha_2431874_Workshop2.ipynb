##  **Some Helper Function:**

### Softmax Function:

```python
import numpy as np

def softmax(z):
    """
    Compute the softmax probabilities for a given input matrix.

    Parameters:
    z (numpy.ndarray): Logits (raw scores) of shape (m, n), where
                       - m is the number of samples.
                       - n is the number of classes.

    Returns:
    numpy.ndarray: Softmax probability matrix of shape (m, n), where
                   each row sums to 1 and represents the probability
                   distribution over classes.

    Notes:
    - The input to softmax is typically computed as: z = XW + b.
    - Uses numerical stabilization by subtracting the max value per row.
    """

    # Your Code Here.
    shift_of_z = z - np.max(z, axis=1, keepdims=True)
    z_exp = np.exp(shift_of_z)
    return z_exp / np.sum(z_exp, axis=1, keepdims=True)
    return
```

### Prediction Function:

```python
def predict_softmax(X, W, b):
    """
    Predict the class labels for a set of samples using the trained softmax model.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d), where n is the number of samples and d is the number of features.
    W (numpy.ndarray): Weight matrix of shape (d, c), where c is the number of classes.
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    numpy.ndarray: Predicted class labels of shape (n,), where each value is the index of the predicted class.
    """
    z = np.dot(X, W) + b
    y_prediction = softmax(z)

    predicted_classes = np.argmax(y_prediction, axis=1)
    return predicted_classes

    return predicted_classes
```

### Loss Function:

```python
def loss_softmax(y_pred, y):
    """
    Compute the cross-entropy loss for a single sample.

    Parameters:
    y_pred (numpy.ndarray): Predicted probabilities of shape (c,) for a single sample,
                             where c is the number of classes.
    y (numpy.ndarray): True labels (one-hot encoded) of shape (c,), where c is the number of classes.

    Returns:
    float: Cross-entropy loss for the given sample.
    """

    epsi = 1e-12
    y_prediction = np.clip(y_pred, epsi, 1.0 - epsi)
    n = y.shape[0]
    cross_entropy_loss = -np.sum(y * np.log(y_prediction)) / n
    return cross_entropy_loss
```

### Cost Function:

```python
def cost_softmax(X, y, W, b):
    """
    Compute the average softmax regression cost (cross-entropy loss) over all samples.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d), where n is the number of samples and d is the number of features.
    y (numpy.ndarray): True labels (one-hot encoded) of shape (n, c), where n is the number of samples and c is the number of classes.
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    float: Average softmax cost (cross-entropy loss) over all samples.
    """

    n = X.shape[0]
    z = np.dot(X, W) + b
    y_prediction = softmax(z)
    total_cost = loss_softmax(y_prediction, y)
    return total_cost

    # Return average loss
    return total_loss / n
```

### Computing Gradients:

```python
def compute_gradient_softmax(X, y, W, b):
    """
    Compute the gradients of the cost function with respect to weights and biases.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d).
    y (numpy.ndarray): True labels (one-hot encoded) of shape (n, c).
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).

    Returns:
    tuple: Gradients with respect to weights (d, c) and biases (c,).
    """


    n, d = X.shape
    z = np.dot(X, W) + b
    y_prediction = softmax(z)
    grad_W = np.dot(X.T, (y_prediction - y)) / n
    grad_b = np.sum(y_prediction - y, axis=0) / n
    return grad_W, grad_b
```

### Implementing Gradient Descent:

```python
def gradient_descent_softmax(X, y, W, b, alpha, n_iter, show_cost=False):
    """
    Perform gradient descent to optimize the weights and biases.

    Parameters:
    X (numpy.ndarray): Feature matrix of shape (n, d).
    y (numpy.ndarray): True labels (one-hot encoded) of shape (n, c).
    W (numpy.ndarray): Weight matrix of shape (d, c).
    b (numpy.ndarray): Bias vector of shape (c,).
    alpha (float): Learning rate.
    n_iter (int): Number of iterations.
    show_cost (bool): Whether to display the cost at intervals.

    Returns:
    tuple: Optimized weights, biases, and cost history.
    """
    cost_history = []
    for i in range(n_iter):
        grad_W, grad_b = compute_gradient_softmax(X, y, W, b)
        W -= alpha * grad_W
        b -= alpha * grad_b
        cost = cost_softmax(X, y, W, b)
        cost_history.append(cost)
        if show_cost and (i % 100 == 0 or i == n_iter - 1):
            print(f"Iteration {i}: Cost = {cost:.6f}")

    return W, b, cost_history
```

## Preparing Dataset:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

def load_and_prepare_mnist(csv_file, test_size=0.2, random_state=42):
    """
    Reads the MNIST CSV file, splits data into train/test sets, and plots one image per class.

    Arguments:
    csv_file (str)       : Path to the CSV file containing MNIST data.
    test_size (float)    : Proportion of the data to use as the test set (default: 0.2).
    random_state (int)   : Random seed for reproducibility (default: 42).

    Returns:
    X_train, X_test, y_train, y_test : Split dataset.
    """

    # Load dataset
    df = pd.read_csv(csv_file)

    # Separate labels and features
    y = df.iloc[:, 0].values  # First column is the label
    X = df.iloc[:, 1:].values  # Remaining columns are pixel values

    # Normalize pixel values (optional but recommended)
    X = X / 255.0  # Scale values between 0 and 1

    # Split data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    # Plot one sample image per class
    plot_sample_images(X, y)

    return X_train, X_test, y_train, y_test

def plot_sample_images(X, y):
    """
    Plots one sample image for each digit class (0-9).

    Arguments:
    X (np.ndarray): Feature matrix containing pixel values.
    y (np.ndarray): Labels corresponding to images.
    """

    plt.figure(figsize=(10, 4))
    unique_classes = np.unique(y)  # Get unique class labels

    for i, digit in enumerate(unique_classes):
        index = np.where(y == digit)[0][0]  # Find first occurrence of the class
        image = X[index].reshape(28, 28)  # Reshape 1D array to 28x28

        plt.subplot(2, 5, i + 1)
        plt.imshow(image, cmap='gray')
        plt.title(f"Digit: {digit}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

# Linear Seperability and Logistic Regression:

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification, make_circles
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

np.random.seed(42)

features_separable, labels_separable = make_classification(
    n_samples=200, n_features=2, n_informative=2,
    n_redundant=0, n_clusters_per_class=1, random_state=42
)

features_sep_train, features_sep_test, labels_sep_train, labels_sep_test = train_test_split(
    features_separable, labels_separable, test_size=0.2, random_state=42
)

clf_separable = LogisticRegression()
clf_separable.fit(features_sep_train, labels_sep_train)

features_circular, labels_circular = make_circles(
    n_samples=200, noise=0.1, factor=0.5, random_state=42
)

features_circ_train, features_circ_test, labels_circ_train, labels_circ_test = train_test_split(
    features_circular, labels_circular, test_size=0.2, random_state=42
)

clf_circular = LogisticRegression()
clf_circular.fit(features_circ_train, labels_circ_train)

def plot_decision_boundary(axes, classifier, data_points, target_labels, plot_title,
                           bg_cmap=plt.cm.RdYlBu, scatter_cmap=plt.cm.coolwarm):
    mesh_step = .02
    x_lower, x_upper = data_points[:, 0].min() - 1, data_points[:, 0].max() + 1
    y_lower, y_upper = data_points[:, 1].min() - 1, data_points[:, 1].max() + 1
    grid_x, grid_y = np.meshgrid(np.arange(x_lower, x_upper, mesh_step), np.arange(y_lower, y_upper, mesh_step))
    predictions = classifier.predict(np.c_[grid_x.ravel(), grid_y.ravel()])
    predictions = predictions.reshape(grid_x.shape)
    axes.contourf(grid_x, grid_y, predictions, alpha=0.6, cmap=bg_cmap)
    scatter = axes.scatter(data_points[:, 0], data_points[:, 1], c=target_labels,
                           edgecolors='k', cmap=scatter_cmap, s=60, linewidths=0.8)
    axes.set_title(plot_title, fontsize=13, fontweight='bold', color='darkblue')
    axes.set_xlabel('Feature 1', fontsize=11)
    axes.set_ylabel('Feature 2', fontsize=11)
    axes.tick_params(axis='both', labelsize=9)
    plt.colorbar(scatter, ax=axes)

fig, subplot_axes = plt.subplots(2, 2, figsize=(14, 11))
fig.patch.set_facecolor('#f5f5f5')

plot_decision_boundary(subplot_axes[0, 0], clf_separable, features_sep_train, labels_sep_train,
    'Linearly Separable Data (Training)',
    bg_cmap=plt.cm.RdYlBu, scatter_cmap=plt.cm.plasma)

plot_decision_boundary(subplot_axes[0, 1], clf_separable, features_sep_test, labels_sep_test,
    'Linearly Separable Data (Testing)',
    bg_cmap=plt.cm.RdYlBu, scatter_cmap=plt.cm.plasma)

plot_decision_boundary(subplot_axes[1, 0], clf_circular, features_circ_train,
    labels_circ_train, 'Non-Linearly Separable Data (Training)',
    bg_cmap=plt.cm.PiYG, scatter_cmap=plt.cm.viridis)

plot_decision_boundary(subplot_axes[1, 1], clf_circular, features_circ_test,
    labels_circ_test, 'Non-Linearly Separable Data (Testing)',
    bg_cmap=plt.cm.PiYG, scatter_cmap=plt.cm.viridis)

plt.suptitle('Decision Boundaries - Logistic Regression', fontsize=15,
             fontweight='bold', color='darkred', y=1.01)
plt.tight_layout()
plt.savefig('decision_boundaries.png', dpi=150, bbox_inches='tight')
plt.show()
```

#Question - 3: Describe any challenges you faced while implementing the code above.

1. Numerical Stability in the Softmax Function
The softmax function involves computing exponentials, which can overflow when input values are large. This was addressed using numerical stabilization subtracting the row-wise maximum from the input before computing the exponential.
2. One-Hot Encoding
The raw MNIST labels are integers (0–9), but the softmax loss function requires one-hot encoded labels. The code explicitly checks for this and applies OneHotEncoder when needed a step that's easy to miss and can cause dimension mismatch errors.
3. Gradient Descent Convergence
Choosing the right learning rate (alpha = 0.1) and number of iterations (1000) required care. Too high a learning rate causes divergence; too low makes training impractically slow. The cost history plot confirms that the chosen hyperparameters lead to smooth and stable convergence.
4. Scalability of Custom Implementation
The custom gradient descent operates on the full training set (48,000 samples × 784 features) in each iteration. This batch gradient descent approach is computationally expensive compared to mini-batch SGD, making training slower a notable challenge in practical settings.
5. Limitation of Linear Models on Non-Linear Data
As illustrated in the decision boundary plots, logistic regression with a softmax output is inherently a linear classifier. It cannot separate classes that require a curved boundary (like the circle dataset), which highlights a key conceptual challenge: choosing the right model architecture for the problem at hand.
